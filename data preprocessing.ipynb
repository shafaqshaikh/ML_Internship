```python
import matplotlib as plt
dataset.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
bmi_med=dataset['BMI'].median()
dataset['BMI']=dataset['BMI'].replace(to_replace=0, value=bmi_med)

bp_med=dataset['BloodPressure'].median()
dataset['BloodPressure']=dataset['BloodPressure'].replace(to_replace=0, value=bp_med)

glu_med=dataset['Glucose'].median()
dataset['Glucose']=dataset['Glucose'].replace(to_replace=0, value=glu_med)

skin_med=dataset['SkinThickness'].median()
dataset['SkinThickness']=dataset['SkinThickness'].replace(to_replace=0, value=skin_med)

in_med=dataset['Insulin'].median()
dataset['Insulin']=dataset['Insulin'].replace(to_replace=0, value=in_med)

dataset.hist(bins=50, figsize=(20,15))
plt.show()







```

```python
print(dataset.head())
dataset.to_csv("preprocess.csv")
```
